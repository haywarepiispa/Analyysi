```python
#--------------------------------------------------------------------------------------#

    # Exercise 1

# variables (list) / arvot (lista)

furniture = ["Chair", "Lamp", "Bed", "Gaming-Chair", "Sofa"]

# print for the whole list / koko listan printti

print("Full list:", furniture)

# way to print 2 first elements / tapa printata listan 2 ensimmäistä elementtiä

print("First two elements: ", furniture[:2])

# if sofa is found on the list, print / jos sohva löytyy listasta, printtaa

for item in furniture:
    if item == "Sofa":
        print("Found: ", item)

#--------------------------------------------------------------------------------------#

    # Exercise 2
import random

# variables (list) / arvot (lista)

thrownDiceNumbers = []

# random numbers / satunnaiset numerot

for i in range(5):
    thrownDiceNumbers.append(random.randint(1, 6))

# print whole list / printtaa koko lista

print("dise numbers: ", thrownDiceNumbers)

# sum / summa

print("Sum of dice numbers: ", sum(thrownDiceNumbers))

# highest value / korkein numero arvo

print("Highest dice number: ", max(thrownDiceNumbers))

#--------------------------------------------------------------------------------------#
```
